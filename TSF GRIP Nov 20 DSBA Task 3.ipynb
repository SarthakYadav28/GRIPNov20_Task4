```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import seaborn as sb
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

### Verifying Data Integrity

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

### Univariate Analysis

```python
df.dtypes
```

```python
df['Discount'].plot.hist()
```

```python
df['Profit'].plot.hist()
```

```python
df['Ship Mode'].value_counts().plot.bar()
```

```python
plt.figure(figsize=(20,8))
plt.plot(df['Profit'])
plt.grid(alpha=0.6)
```

```python
plt.figure(figsize=(20,8))
sb.countplot(df['State'])
plt.xticks(fontsize=16,rotation=90)
plt.ylabel('Frequency',fontsize=20)
plt.xlabel('States',fontsize=20)
plt.grid(alpha=0.6)
plt.show()
```

```python
sb.countplot('Category', data=df)
```

```python
plt.figure(figsize=(20,8))
sb.countplot('Sub-Category', data=df)
```

```python
plt.figure(figsize=(20,8))
sb.countplot('Category', hue='Sub-Category', data=df)
```

```python
sb.countplot('Region', data=df)
```

```python
plt.figure(figsize=(20,15))
sb.countplot('Region', hue='State', data=df)
```

### Bivariate Analysis

```python
sb.jointplot(x='Profit', y='Discount', data=df)
```

```python
sb.jointplot(x='Profit', y='Quantity', data=df)
```

```python
correlation=df.corr()
sb.heatmap(correlation,annot=True)
```

```python
sb.pairplot(df, hue='Category', diag_kind='kde', kind='scatter')
```

```python
df2=df.drop(['Postal Code'], axis=1)
```

```python
plt.figure(figsize=(20, 8))
sb.boxplot(data=df2)
```
